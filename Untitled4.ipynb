```python
import pandas as pd
import zipfile
import matplotlib.pyplot as plt

class InsuranceAnalyzer:
    def __init__(self, zip_path, csv_path):
        with zipfile.ZipFile(zip_path) as z:
            with z.open(csv_path) as f:
                self.df = pd.read_csv(f)

    def average_age(self):
        return self.df['age'].mean()

    def average_age_with_children(self):
        return self.df[self.df['children'] >= 1]['age'].mean()

    def region_distribution(self):
        return self.df['region'].value_counts()

    def average_cost_by_smoking(self):
        return self.df.groupby('smoker')['charges'].mean()

    def show_charges_by_smoker(self):
        avg_costs = self.df.groupby('smoker')['charges'].mean()
        avg_costs.plot(kind='bar', color=['blue', 'red'])
        plt.title("Average Medical Charges: Smokers vs Non-Smokers")
        plt.xlabel("Smoker")
        plt.ylabel("Average Charges")
        plt.xticks(rotation=0)
        plt.tight_layout()
        plt.show()
```
