```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
import os
%matplotlib inline
```

```python
life_exp = pd.read_csv("all_data.csv")
life_exp.head()
```

```python
print(life_exp["Country"].value_counts().index)
```

```python
print(life_exp["Year"].value_counts().index.tolist())
```

```python
plt.scatter(life_exp["GDP"], life_exp["Life expectancy at birth (years)"])
plt.xlabel("GDP")
plt.ylabel("Life Expectancy at Birth (years)")
plt.title("Life Expectancy vs GDP")
plt.show()
```

```python
plt.scatter(life_exp["GDP"], life_exp["Life expectancy at birth (years)"])
plt.xscale("log")   # because GDP has variations 
plt.xlabel("GDP (log scale)")
plt.ylabel("Life Expectancy at Birth (years)")
plt.title("Life Expectancy vs GDP (log scale)")
plt.show()
```

```python
sns.scatterplot(
    data=life_exp,
    x="GDP",
    y="Life expectancy at birth (years)",
    hue="Country"  
)
# to clarity of GDP distribution
plt.xscale("log")  
plt.title("Life Expectancy vs GDP by Country")
plt.show()
```

```python
sns.lineplot(data=life_exp, x="Year", y="Life expectancy at birth (years)", hue="Country")
plt.title("Life Expectancy Over Time by Country")
plt.show()


sns.lineplot(data=life_exp, x="Year", y="GDP", hue="Country")
plt.yscale("log")
plt.title("GDP Over Time by Country")

# Moveing legend outside the plot to be more readable
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.tight_layout()
plt.show()



```

```python
# GDP distribution using Kernel Density Estimation.
sns.kdeplot(life_exp["GDP"], fill=True)  
plt.title("GDP Distribution")
plt.show()

# Life Expectancy distribution using Kernel Density Estimation.
sns.kdeplot(life_exp["Life expectancy at birth (years)"], fill=True, color="green")
plt.title("Life Expectancy Distribution")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="Country", y="GDP", data=life_exp)
plt.yscale("log")  
plt.title("GDP Distribution by Country (log scale)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x="Country", y="Life expectancy at birth (years)", data=life_exp, color="green")
plt.title("Life Expectancy Distribution by Country")
plt.xticks(rotation=45)
plt.show()


# The violin plot of the life expectancy distribution by country
os.makedirs("plots", exist_ok=True)
print(os.listdir("plots"))

plt.figure(figsize=(8,5))
sns.violinplot(
    data=life_exp,
    x="Country",
    y="Life expectancy at birth (years)",
    hue="Country",
    palette="Set2",
    legend=False
)
plt.title("Life Expectancy Distribution by Country")
plt.ylabel("Life Expectancy at Birth (years)")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.show()
plt.savefig("plots/life_expectancy_violin.png", dpi=300, bbox_inches='tight')
plt.show()

```

```python
# Creating color palette
plots = sns.color_palette("tab10", n_colors=len(life_exp["Country"].unique()))

# Creating FacetGrid
facet_plot = sns.FacetGrid(life_exp, col="Country", col_wrap=4, height=3, sharey=False)

# Plot a line for each country using the same FacetGrid object
for ax, (country, data) in zip(facet_plot.axes.flat, life_exp.groupby("Country")):
    sns.lineplot(
        data=data,
        x="Year",
        y="GDP",
        ax=ax,
        color=plots[list(life_exp["Country"].unique()).index(country)],
        marker="o"
    )
    ax.set_title(country)

# Giving labels and overall title
facet_plot.set_axis_labels("Year", "GDP")
plt.subplots_adjust(top=0.9)
facet_plot.fig.suptitle("GDP Over Time by Country (Different Colors per Country, Independent Y-Axis)")

# Saveing the plot
facet_plot.savefig("plots/GDP_FacetGrid.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
# Creating color palette
plots = sns.color_palette("tab10", n_colors=len(life_exp["Country"].unique()))

# Creating FacetGrid
facet_plot = sns.FacetGrid(life_exp, col="Country", col_wrap=4, height=3, sharex=False, sharey=False)

# Plot scatter + line in each subplot with a unique color
for ax, (country, data) in zip(facet_plot.axes.flat, life_exp.groupby("Country")):
    sns.scatterplot(
        data=data,
        x="GDP",
        y="Life expectancy at birth (years)",
        ax=ax,
        color=plots[list(life_exp["Country"].unique()).index(country)],
        s=50
    )
    sns.lineplot(
        data=data.sort_values("GDP"),
        x="GDP",
        y="Life expectancy at birth (years)",
        ax=ax,
        color=plots[list(life_exp["Country"].unique()).index(country)]
    )
    ax.set_title(country)

# Giving labels and overall title
facet_plot.set_axis_labels("GDP", "Life Expectancy at Birth (years)")
plt.subplots_adjust(top=0.9)
facet_plot.fig.suptitle("GDP vs Life Expectancy at Birth by Country (Different Colors, Independent Axes)")

# Save the plot 
facet_plot.savefig("plots/GDP_vs_LifeExpectancy_FacetGrid.png", dpi=300, bbox_inches='tight')

plt.show()
```

Conclusion :


Despite the dataset being small, the project revealed several interesting insights through visualizations. Life expectancy in most countries falls in the mid-to-high 70s, although Zimbabwe remains notably lower at around 50 years. Over time, both GDP and life expectancy have shown upward trends, with China experiencing the strongest economic growth and Zimbabwe showing the largest improvement in life expectancy. The relationship between GDP and life expectancy is clearly positive, suggesting that higher economic prosperity is associated with better health outcomes. The distribution of life expectancy is slightly left-skewed, and the distribution of GDP is gently right-skewed, indicating that most observations are concentrated at the higher end. Overall, the analysis highlights patterns of economic and health development across the six nations and demonstrates that even a small dataset can provide meaningful insights.
